```python

df.Date = pd.to_datetime(df.Date)
df.set_index('Date', inplace=True)
```

```python

df['Log Ret'] = np.log(df['Adj Close']) - np.log(df['Adj Close'].shift(1))

monthly = df.resample('BM').apply(lambda x: x[-1]) # use BA for business annual (year)
monthly
```
