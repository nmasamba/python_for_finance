```python

import matplotlib.pyplot as plt
```

```python

# Derived manually from Yahoo Finance
time = [1/12, 2/12, 3/12, 6/12, 1, 2, 3, 5, 7, 10, 20, 30]
rates = [2.21, 2.23, 2.33, 2.48, 2.67, 2.89, 2.95, 3.01, 3.10, 3.17, 3.29, 3.37]
```

```python

plt.plot(time, rates)
plt.title('Term Structure of Interest Rates')
plt.xlabel('Time')
plt.ylabel('Risk-Free Rate')
plt.show()
```


The upward sloping term structure means the yield on longer term securities is higher than on shorter term securities.
