Retrieve text data from text document for manipulation in notebook.

```python

import numpy as np
import pandas as pd
```

```python
file=open("market_data/F-F_Research_Data_Factors.txt","r")
data=file.readlines()

data
```

```python
f=[]
index=[]
for i in range(1,np.size(data)):
    t=data[i].split()
    index.append(int(t[0]))
    for j in range(1,5):
        k=float(t[j])
        f.append(k/100)
        
```

```python

f = np.array(f)
n = len(f)
n
```

```python

f1 = np.reshape(f, [int(n/4), 4])
ff = pd.DataFrame(f1, index=index, columns=['Mkt-Rf', 'SMB', 'HML', 'RF'])
```
